```python
from matplotlib import pyplot as plt
```

```python
!wget "https://raw.githubusercontent.com/yamnor/yamlab-note/main/ethylene-conical/conical-spnflp.log"
```

```python
energy = []
for n in range(4):
  energy.append([float(line.split()[2]) for line in open('conical-spnflp.log', 'r').readlines() if ' %d  A '%(n) in line])
```

```python
step = list(range(len(energy[0])))
fig, ax = plt.subplots()
ax.plot(step, energy[0], label="State 0")
ax.plot(step, energy[1], label="State 1")
ax.plot(step, energy[2], label="State 2")
ax.plot(step, energy[3], label="State 3")
ax.set_xlabel("Step")
ax.set_ylabel("Energy [au]")
ax.legend()
plt.show()
```

```python
s_squared = []
for n in range(1,4):
  s_squared.append([float(line.split()[4]) for line in open('conical-spnflp.log', 'r').readlines() if ' %d  A '%(n) in line])
```

```python
step = list(range(len(s_squared[0])))
fig, ax = plt.subplots()
ax.plot(step, s_squared[0], label="State 1")
ax.plot(step, s_squared[1], label="State 2")
ax.plot(step, s_squared[2], label="State 3")
ax.set_xlabel("Step")
ax.set_ylabel("$<S^2>$")
ax.legend()
plt.show()
```
