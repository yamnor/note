```python
from matplotlib import pyplot as plt
```

```python
energy = []
for n in range(4):
  energy.append([float(line.split()[2]) for line in open('conical-spinflip.log', 'r').readlines() if ' %d  A '%(n) in line])
```

```python
step = list(range(len(energy[0])))
fig, ax = plt.subplots()
ax.plot(step, energy[2], label="S0")
ax.plot(step, energy[3], label="S1")
ax.set_xlabel("Step")
ax.set_ylabel("Energy [au]")
ax.legend()
plt.show()
```

```python
s_squared = []
for n in range(1,4):
  s_squared.append([float(line.split()[4]) for line in open('conical-spinflip.log', 'r').readlines() if ' %d  A '%(n) in line])
```

```python
step = list(range(len(s_squared[0])))
fig, ax = plt.subplots()
ax.plot(step, s_squared[1], label="S0")
ax.plot(step, s_squared[2], label="S1")
ax.set_xlabel("Step")
ax.set_ylabel("$<S^2>$")
ax.legend()
plt.show()
```
