```python
from matplotlib import pyplot as plt
```

```python
!wget "https://raw.githubusercontent.com/yamnor/yamlab-note/main/ethylene-conical/optimize-excite.log"
```

```python
energy = []
for n in range(4):
  energy.append([float(line.split()[2]) for line in open('optimize-excite.log', 'r').readlines() if ' %d  A '%(n) in line])
```

```python
step = list(range(len(energy[0])))
fig, ax = plt.subplots()
ax.plot(step, energy[0], label="S0")
ax.plot(step, energy[1], label="S1")
ax.set_xlabel("Step")
ax.set_ylabel("Energy [au]")
ax.legend()
plt.show()
```
