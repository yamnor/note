```python
from matplotlib import pyplot as plt
```

```python
energy = []
for n in range(4):
  energy.append([float(line.split()[2]) for line in open('output.txt', 'r').readlines() if ' %d  A '%(n) in line])
```

```python
x = list(range(len(energy[0])))
fig, ax = plt.subplots()
ax.plot(x, energy[0])
ax.plot(x, energy[1])
```
